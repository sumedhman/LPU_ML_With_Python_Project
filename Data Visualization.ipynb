## Data Visualization 

```python
# importing required modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 50)
print (x, type(x), len(x))
```

```python
sinus = np.sin(x)
print (sinus, type(sinus), len(sinus))
```

```python
plt.plot(x, sinus)
plt.show()
```

```python
plt.plot(x, sinus, "o")
plt.show()
```

```python
plt.plot(x, sinus, "--")
plt.show()
```

```python
plt.plot(x, sinus, "*")
plt.show()
```

```python
plt.plot(x, sinus, "-")
plt.show()
```

```python
help (plt.plot)
```

```python
# Rapid multiplot
cosinus = np.cos(x)
plt.plot(x, sinus, "-b", x, sinus, "ob", x, cosinus, "-r", x, cosinus, "or")
plt.xlabel("Values of x...")
plt.ylabel("Sin(x) / Cos(x)")
plt.title("Plotting Sin(x) and Cos(x) vs. x values...")
plt.show()
```

```python
plt.plot(x, sinus, label='Sin(x)', color='blue', linestyle='--', linewidth=2)
plt.plot(x, cosinus, label='Cos(x)', color='red', linestyle='-', linewidth=2)
plt.xlabel("All x-values...")
plt.ylabel("All y-values...")
plt.title("Plotiing Six(x) and Cos(x) against x-values...")
plt.legend(loc = 0)
plt.show()
```

### Scatter Plot

```python
import pandas as pd
try:
    salary = pd.read_csv('salary_table.csv')
    print (salary.shape)
except:
    print ("Error opening the file...")
salary.head(10)
```

### Simple Scatter with Colors

```python
colors_edu = {'Bachelor':'r', 'Master':'g', 'Ph.D':'blue'}
print (colors_edu, type(colors_edu))
```

```python
df = salary.copy()
plt.scatter(df['experience'], df['salary'], c=df['education'].apply(lambda x:colors_edu[x]), s=50)
plt.show()
```

### Scatter Plot with Colors and Symbols

```python
for values, data in df.groupby(['education', 'management']):
    # print (values, ":\n", data)
    edu, manager = values
    print (f"edu = {edu} and manager = {manager}...")
    print (data)
```

```python
# define colors / symbols manually
symbols_mang = dict(Y="*", N=".")
colors_edu = {'Bachelor':'r', 'Master':'g', 'Ph.D':'blue'}
print (symbols_mang, type(symbols_mang))
print (colors_edu, type(colors_edu))
```

```python
## Figure size
plt.figure(figsize=(6,5))
# group by education x management => 3 x 2 = 6 groups or combinations
for values, data in df.groupby(['education', 'management']):
    # print (values, ":\n", data)
    edu, manager = values
    plt.scatter(data['experience'], data['salary'], marker=symbols_mang[manager], 
               color=colors_edu[edu], s=150, label=manager + " / " + edu)
# set labels
plt.xlabel('Experience (in years)')
plt.ylabel('Salary (in USD)')
plt.title('Experience (in years) vs. Salary (in USD)')
plt.legend(loc=4)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```
